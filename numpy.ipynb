```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([70,50,10,80])
my_label = ["BMW","Jeep","Toyota","Volvo"]
my_colors = ["blue","purple","green","red"]
my_explode = [0.1,0,0,0]
plt.pie(x,labels=my_label, colors=my_colors, explode=my_explode, shadow=True)
plt.legend(title="cars")

plt.show()
```
